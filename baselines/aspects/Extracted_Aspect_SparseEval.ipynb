```python
from Aspect_Sparse_Baselines import OWC, TFIDF, BM25
import numpy as np
import json
from helper import *
```

```python
config = load_config()
with open(config['data_path']) as f:
  all_data = json.load(f)

with open(config['folds_path'], 'r', encoding='utf-8') as f:
  folds = json.load(f)

aspects_path = "bloom-3b_extracted_aspects_17.json"
with open(aspects_path, 'r', encoding='utf-8') as f:
  aspects_dict = json.load(f)

train_inds = folds[0][0]
test_inds = folds[0][2]

test_data = [all_data[i] for i in test_inds]
```

```python
def custom_gmean(lst):
  gmean = np.prod(lst)**(1/len(lst))
  return gmean

agg_fcn = custom_gmean
```

```python
OWC_baseline = OWC(test_data)
score_fcn = OWC_baseline.owc_score

OWC_baseline.clean_data()
OWC_baseline.get_results(score_fcn, agg_fcn, aspects_dict)
```

```python
TFIDF_baseline = TFIDF(test_data)
score_fcn = TFIDF_baseline.tfidf_score

cleaned_descriptions = TFIDF_baseline.clean_data()
TFIDF_baseline.calc_df(cleaned_descriptions)
TFIDF_baseline.get_results(score_fcn, agg_fcn, aspects_dict)
```

```python
BM25_baseline = BM25(test_data)
score_fcn = BM25_baseline.bm25_score

cleaned_descriptions = BM25_baseline.clean_data()
BM25_baseline.fit(cleaned_descriptions)
BM25_baseline.get_results(score_fcn, agg_fcn, aspects_dict)
```
